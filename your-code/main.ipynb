#### Display the top 10 languages by number of native speakers stored in a pandas dataframe.

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
````

````python
# your code here

languages_table = soup.find_all('table', attrs = {'class':'wikitable sortable'})[0]

table_head = [cell.text.replace('\n',' ') for cell in languages_table.find_all('th')]

rows = [cell.text.replace('\n',' ') for cell in languages_table.find_all('td')]

dictionary = {table_head[i]: rows[i::len(table_head)] for i in range(len(table_head))} # 0 -- 1 -- 2 ---3 until the end of the row, jumps
dictionary

df = pd.DataFrame(dictionary)
df.head(10)
````

# Display IMDB's top 250 data (movie name, initial release, director name and stars) as a pandas dataframe.

````python
# This is the url you will scrape in this exercise 
url = 'https://www.imdb.com/chart/top'
````

````python
# your code here

#imdb = pd.DataFrame(list(zip(movies, year, directors, star1, star2)), columns= ['movies', 'release', 'directors', 'star1', 'star2'])\n",


import re 

html = requests.get(url).content
soup = BeautifulSoup(html, 'lxml')

table = soup.find_all('table' , attrs= {'class':'chart full-width'})[0]


row = table.find_all('tr')
row

all_cast = [cast.get('title').split(',') for cast in table.find_all('a') if cast.get('title') is not None]
director = [name[0] for name in all_cast]
director

stars = [','.join(name[1:]).strip() for name in all_cast]
stars


rows = [r.text.replace('\n', '').replace('\xa0','').strip() for r in row]
rows = rows[1:]

rows

title = [' '.join(re.findall('\w*', name)[1:-4]) for name in rows ]
title = [name[1:-4] for name in title]



year = [re.findall('\d{4} ', numb)[0] for numb in title ]


title_name = [x[6:-30] for x in title]
title_name

df = pd.DataFrame(title_name)
df['year'] = year
df['Director'] = director
df['Stars'] = stars

df = df.rename(columns = {0 : 'Title'})

df.tail()
````

#### Find the book name, price and stock availability as a pandas dataframe.

````python
# This is the url you will scrape in this exercise. 
# It is a fictional bookstore created to be scraped. 
url = 'http://books.toscrape.com/'

html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
````

````python
# your code here

article = soup.find_all('article', attrs = {'class':'product_pod'})

#title

title = [art.find('h3').text for art in article]

#price 

price = [art.find('p', attrs = {'class':'price_color'}).text for art in article]

#stock
stock = [art.find('p', attrs = {'class':'instock availability'}).text.replace('\n', '').strip() for art in article]

data = pd.DataFrame({'Title':title, 'Price':price, 'Stock':stock})
data
````
